# Membangun Recommendation System

```python
from sklearn.metrics.pairwise import cosine_similarity

cos_similarity = cosine_similarity(dataset[numerical_columns])
cos_similarity
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
dataset['track_name_encoded'] = label_encoder.fit_transform(dataset['track_name'])
```

```python
dataset['track_name_encoded'].nunique()
```

```python
def recomendation_song(input_song, similarity_data=cos_similarity, top_k=5, only_same_genre=False):
    song_index = dataset[dataset['track_name'] == input_song].index
    if len(song_index) == 0:
        print("Lagu tidak ditemukan dalam dataset.")
        return None

    genre_input = dataset.loc[song_index[0], 'playlist_genre']
    track_input = dataset.loc[song_index[0], 'track_name_encoded']
    distances = similarity_data[track_input]
    songs_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:]

    songs_recommendation = pd.DataFrame(columns=['Artist', 'Song', 'Genre'])
    for idx, song in songs_list:
        if only_same_genre:
            if dataset.loc[idx, 'playlist_genre'] != genre_input:
                continue
        row = pd.DataFrame([{'Artist': dataset.loc[idx, 'track_artist'], 'Song': dataset.loc[idx, 'track_name'], 'Genre': dataset.loc[idx, 'playlist_genre']}])
        songs_recommendation = pd.concat([songs_recommendation, row], ignore_index=True)
        if songs_recommendation.shape[0] == top_k:
            break

    return songs_recommendation
```

# Test Recommendation
